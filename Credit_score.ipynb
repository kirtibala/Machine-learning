```python
x=df.drop('DEFAULT', axis = 1)
y=df['DEFAULT']
print('x',x)
print('y', y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,precision_score,accuracy_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
rand_fost = RandomForestClassifier(random_state=30)
rand_fost.fit(x_train, y_train)
```

```python
y_pred_train = rand_fost.predict(x_train)
conf_matrix=confusion_matrix(y_pred_train,y_train)
print(conf_matrix)
prec_score=precision_score(y_pred_train,y_train)
print (prec_score)
acu_score= accuracy_score(y_pred_train,y_train)
print(acu_score)
```

```python
y_pred_test = rand_fost.predict(x_test)
conf_matrix=confusion_matrix(y_pred_test,y_test)
print(conf_matrix)
prec_score=precision_score(y_pred_test,y_test)
print (prec_score)
acu_score= accuracy_score(y_pred_test,y_test)
print(acu_score)
```
