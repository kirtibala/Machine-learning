```python
x_scale=scaler.fit_transform(x)
```

```python
x_scale_df= pd.DataFrame(x_scale,columns=x.columns)
x_scale_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scale_df,y,test_size=0.2)
```

```python
x_train
```

```python
df[['hypertension']].boxplot()
```

```python
y_train
```

```python
from sklearn.svm import SVC
```

```python
svm_mod=SVC(C=6,kernel='linear')
svm_mod.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
y_pred_train = svm_mod.predict(x_train)
 
cnf_matrix = confusion_matrix(y_train, y_pred_train)
print("Confusion Matrix :\n",cnf_matrix)
 
acc_score = accuracy_score(y_train, y_pred_train)
print("Accuracy Score is :\n",acc_score)
 
clf_report = classification_report(y_train, y_pred_train)
print("Classification Report :\n",clf_report)
```

```python
y_pred_test = svm_mod.predict(x_test)
 
cnf_matrix = confusion_matrix(y_test, y_pred_train)
print("Confusion Matrix :\n",cnf_matrix)
 
acc_score = accuracy_score(y_test, y_pred_train)
print("Accuracy Score is :\n",acc_score)
 
clf_report = classification_report(y_test, y_pred_train)
print("Classification Report :\n",clf_report)
```
